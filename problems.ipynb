## Problem 5: The Penguins

For the remaining problems, I will use the Palmer Penguins dataset.

The Palmer Penguins dataset contains measurements for three penguin species: Adelie, Chinstrap, and Gentoo. The dataset includes variables such as bill length, bill depth, flipper length, body mass, island, sex, and year.

In this problem, I load the dataset into a pandas DataFrame and use descriptive statistics to summarise the numeric variables.

```python
# Load the Palmer Penguins dataset from a raw CSV URL.
penguins_url = "https://gist.githubusercontent.com/slopp/ce3b90b9168f2f921784de84fa445651/raw/penguins.csv"
penguins = panda.read_csv(penguins_url)
```

```python
# Display the first five rows of the dataset.
penguins.head()
```

```python
# Display basic information about the dataset.
penguins.info()
```

```python
# Calculate descriptive statistics for the numeric variables.
penguins[[
    "bill_length_mm",
    "bill_depth_mm",
    "flipper_length_mm",
    "body_mass_g"
]].describe()
```

The `describe()` method gives descriptive statistics for the numeric variables in the dataset.

The `mean` is the average value. It gives a general idea of the typical measurement for each variable.

The `std` is the standard deviation. It shows how spread out the values are from the mean. A larger standard deviation means the values vary more.

The quartiles show how the data is divided into sections. The `25%` value is the first quartile, meaning 25% of the values are below it. The `50%` value is the median, meaning half the values are below it and half are above it. The `75%` value is the third quartile, meaning 75% of the values are below it.

Together, these statistics help describe the size, spread, and distribution of the penguin measurements.

## Problem 6: Visualizing Distributions

For this problem, I created visualisations to show the distribution of variables in the Palmer Penguins dataset.

A distribution shows how the values of a variable are spread out. For numeric variables, histograms are useful because they show how often different ranges of values occur. For categorical variables, bar charts are useful because they show the number of observations in each category.

To make the charts easier to read, I used descriptive labels for the variable names. For example, `bill_length_mm` is displayed as `Bill Length (mm)` and `body_mass_g` is displayed as `Body Mass (g)`.

For the categorical variables, I also formatted the category names so that values such as `male` and `female` are displayed as `Male` and `Female` on the charts.

```python
# Create histograms for the numeric variables in the Palmer Penguins dataset. 

numeric_columns = {
    "bill_length_mm": "Bill Length (mm)",
    "bill_depth_mm": "Bill Depth (mm)",
    "flipper_length_mm": "Flipper Length (mm)",
    "body_mass_g": "Body Mass (g)"
}

for column, label in numeric_columns.items():
    plot.figure(figsize=(6, 4))
    plot.hist(penguins[column].dropna(), bins=20)
    plot.title(f"Distribution of {label}")
    plot.xlabel(label)
    plot.ylabel("Frequency")
    plot.tight_layout()
    plot.show()
```

The histograms show the distributions of the numeric variables in the dataset.

The `Bill Length (mm)` histogram shows the spread of penguin bill lengths. The values are not all centred around one single point, which may be because the dataset contains three different penguin species.

The `Bill Depth (mm)` histogram shows the spread of bill depths. This variable also appears to have more than one cluster of values.

The `Flipper Length (mm)` histogram shows variation in penguin flipper lengths. Some penguins have noticeably longer flippers than others.

The `Body Mass (g)` histogram shows the distribution of penguin body masses. The values are spread over a wide range, showing that some penguins are much heavier than others.

```python
# Create bar charts for categorical variables in the Palmer Penguins dataset.

categorical_columns = {
    "species": "Species",
    "island": "Island",
    "sex": "Sex"
}

for column, label in categorical_columns.items():
    plot.figure(figsize=(6, 4))

    # Capitalise the category names for display on the chart.
    counts = penguins[column].dropna().str.capitalize().value_counts()

    counts.plot(kind="bar")

    plot.title(f"Distribution of {label}")
    plot.xlabel(label)
    plot.ylabel("Count")
    plot.xticks(rotation=45)
    plot.tight_layout()
    plot.show()
```

The bar charts show the distributions of the categorical variables.

The `Species` chart shows how many penguins belong to each species. This helps show whether the dataset is balanced across Adelie, Chinstrap, and Gentoo penguins.

The `Island` chart shows how many penguins were recorded on each island.

The `Sex` chart shows the number of male and female penguins in the dataset. The chart labels are capitalised to make them easier to read. Some values may be missing because not every penguin has a recorded sex value.
