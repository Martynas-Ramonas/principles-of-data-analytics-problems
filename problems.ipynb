## Problem 6: Visualizing Distributions

For this problem, I created two plots to show the distribution of body mass in the Palmer Penguins dataset.

I used `body_mass_g` because body mass is an easy measurement to understand and compare.

A histogram shows how often different ranges of values occur. A box plot gives a summary of the data, including the median, spread, and possible outliers.

```python
# Select the numeric variable to investigate.

body_mass = penguins["body_mass_g"].dropna()
```

```python
# Create a histogram of penguin body mass.

plot.figure(figsize=(8, 6))

plot.hist(
    body_mass,
    bins=20
)

plot.title("Distribution of Penguin Body Mass")
plot.xlabel("Body Mass (g)")
plot.ylabel("Frequency")
plot.grid(True)
plot.tight_layout()
plot.show()
```

```python
# Create a box plot of penguin body mass.

plot.figure(figsize=(8, 6))

plot.boxplot(
    body_mass,
    vert=False
)

plot.title("Box Plot of Penguin Body Mass")
plot.xlabel("Body Mass (g)")
plot.tight_layout()
plot.show()
```

The histogram shows the overall shape of the body mass distribution. It shows how often different body mass ranges occur in the dataset.

The box plot gives a more compact summary of the same variable. It shows the median body mass, the spread of the middle 50% of the values, and whether there are any possible outliers.

The histogram is better for seeing the shape of the distribution. The box plot is better for quickly seeing the median, spread, and possible unusual values.

## Problem 7: Investigating Relationships

For this problem, I investigated whether larger penguins tend to have longer bills.

I used a scatter plot to compare body mass and bill length. Body mass is shown on the X-axis and bill length is shown on the Y-axis.

The points are colour-coded by species so that the three penguin species can be compared.

```python
# Create a scatter plot to investigate whether larger penguins tend to have longer bills.

plot.figure(figsize=(8, 6))

plot.scatter(
    penguins["body_mass_g"],
    penguins["bill_length_mm"],
    alpha=0.7
)

plot.title("Relationship Between Body Mass and Bill Length")
plot.xlabel("Body Mass (g)")
plot.ylabel("Bill Length (mm)")
plot.grid(True)
plot.tight_layout()
plot.show()
```

The scatter plot shows the relationship between `Body Mass (g)` and `Bill Length (mm)`.

There appears to be a positive relationship between the two variables. Penguins with a higher body mass often have longer bills.

The points do not form a perfect straight line, so body mass does not perfectly predict bill length. However, the overall pattern suggests that the two measurements are related.

```python
# Create a scatter plot of body mass and bill length, using different colours for each penguin species.

species_colours = {
    "Adelie": "tab:blue",
    "Chinstrap": "tab:orange",
    "Gentoo": "tab:green"
}

plot.figure(figsize=(8, 6))

for species, colour in species_colours.items():
    species_data = penguins[penguins["species"] == species]

    plot.scatter(
        species_data["body_mass_g"],
        species_data["bill_length_mm"],
        label=species,
        alpha=0.7,
        color=colour
    )

plot.title("Relationship Between Body Mass and Bill Length by Species")
plot.xlabel("Body Mass (g)")
plot.ylabel("Bill Length (mm)")
plot.legend(title="Species")
plot.grid(True)
plot.tight_layout()
plot.show()
```

The scatter plot with species shown separately gives more detail about the relationship between body mass and bill length.

The chart shows that species is important. The different species appear in different parts of the chart, which means the relationship between body mass and bill length is not exactly the same for every species.

Gentoo penguins generally have a higher body mass, while Adelie and Chinstrap penguins are generally lighter. Bill length also differs between the species.

This means body mass is useful for understanding bill length, but species also matters.

## Problem 8: Modelling the Data

For this problem, I fitted a straight line to the relationship between body mass and bill length in the Palmer Penguins dataset.

The aim is to investigate whether a penguin's body mass can be used to predict its bill length.

A straight-line model has the form:

`y = mx + c`

where:

- `y` is the predicted bill length.
- `x` is the body mass.
- `m` is the slope of the line.
- `c` is the intercept.

```python
# Select the columns needed for the model.
# Drop missing values so the model only uses complete rows.

model_data = penguins[["body_mass_g", "bill_length_mm"]].dropna()

x = model_data["body_mass_g"]
y = model_data["bill_length_mm"]
```

```python
# Fit a straight line to the data using numpy.polyfit.
# The 1 means we want a straight line.

slope, intercept = numpy.polyfit(x, y, 1)

print(f"Slope: {slope:.4f}")
print(f"Intercept: {intercept:.4f}")
```

The slope and intercept are used to create the straight line.

The slope shows how much the bill length changes as body mass increases.

The intercept is where the line would start if body mass was zero. In this dataset, the intercept is mainly needed to draw the line.

```python
# Use the slope and intercept to calculate the predicted bill length values.

predicted_y = slope * x + intercept
```

```python
# Calculate the R-squared value.
# R-squared shows how well the line fits the data.

residuals = y - predicted_y
sum_squared_residuals = numpy.sum(residuals ** 2)

total_variation = y - numpy.mean(y)
total_sum_of_squares = numpy.sum(total_variation ** 2)

r_squared = 1 - (sum_squared_residuals / total_sum_of_squares)

print(f"R-squared: {r_squared:.4f}")
```

The R-squared value shows how well the straight line fits the data.

If R-squared is close to 1, the line is a very good fit.

If R-squared is close to 0, the line is not a good fit.

This helps show whether body mass is useful for estimating bill length.

```python
# Plot the original data points and the straight-line model.

plot.figure(figsize=(8, 6))

plot.scatter(
    x,
    y,
    alpha=0.7,
    label="Penguins"
)

plot.plot(
    x,
    predicted_y,
    label="Best-Fit Line"
)

plot.title("Best-Fit Line for Body Mass and Bill Length")
plot.xlabel("Body Mass (g)")
plot.ylabel("Bill Length (mm)")
plot.legend()
plot.grid(True)
plot.tight_layout()
plot.show()
```

The chart shows the penguins as points and the straight-line model as a line.

The line shows the general relationship between body mass and bill length.

The chart suggests that heavier penguins often have longer bills. However, the points are not all close to the line, so the model is not perfect.

This means body mass can give us some idea of bill length, but it cannot predict bill length exactly. Other information, such as the penguin species, would probably help make a better prediction.
